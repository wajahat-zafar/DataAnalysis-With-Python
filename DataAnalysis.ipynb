```python
from jupyter_dash import JupyterDash
from dash import Dash
from dash import dcc
from dash import html
import dash_bootstrap_components as dbc
from dash.dependencies import Output, Input
import pandas as pd
import plotly.express as px
import math
from dash import no_update 
```

```python
import plotly.graph_objects as go
```

```python
import pandas as pd
import plotly.express as px
```

```python
external_stylesheet = [dbc.themes.BOOTSTRAP]
app = JupyterDash(__name__, external_stylesheets = external_stylesheet)
 
geo_dropdown = dcc.Dropdown(options=Data['Horsepower'].unique(),
                            value='None')
geo_dropdown2 = dcc.Dropdown(options=Data['FuelType'].unique(),
                            value='None')
app.layout = html.Div([
    html.Div([
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.Label("DashBoard Analysis of Barplot"),
                    
                    geo_dropdown
                  
                    
                ])
            ),
         dbc.Col(
                html.Div([
                    html.Label("DashBoard Analysis of Boxplot"),
                    geo_dropdown2 
                    
                                 
                ])
         )
          
        ])

    ]),
    html.Div([
        dbc.Row([
            dbc.Col(
                html.Div(
                    dcc.Graph(id = 'price_graph'),
            ), style = {'width': '40%', 'align' : 'right', 'display' : 'inline-block'}
        ),
        dbc.Col(
            html.Div(
                dcc.Graph(id ='price_graph2'),
            ),style = {'width': '40%', 'align' : 'right', 'display' : 'inline-block'}
        )

        ])
    ])
])


@app.callback(
    Output(component_id='price_graph', component_property='figure'),
    Input(component_id=geo_dropdown, component_property='value')
)
def chart1(Horsepower_Value):
    Doornumber_Filtered = Data[Data['Horsepower'] == Horsepower_Value]
    bar_fig = px.bar(Doornumber_Filtered,
                       x='CC', y='Price',
                       color='MetColor',
                       title=f'CC and Price groupby Horsepower {Horsepower_Value}')
    return bar_fig

@app.callback(
    Output(component_id='price_graph2', component_property='figure'),
    Input(component_id=geo_dropdown2, component_property='value')
)
def chart2(FuelType_Value):
    Fuel_Filtered = Data[Data['FuelType'] == FuelType_Value]
    box_fig = px.box( Fuel_Filtered,
                       x='Doornumber', y='Price',
                       color='MetColor',
                       title=f'Doornumber and Price groupby FuelType {FuelType_Value}')
    return box_fig

                





        
  


app.run_server(mode='inline')
```

**Explaination:**
The above interpretation is the dashboard analysis of prices of cars in CC and prices of cars with respect to doornumbers. On the left there is a bar plot which is showing the respective variations in prices by changing the horse powers.
Whereas, the box plot with different doornumbers showing the respective variations in the graph by changing the fuel types.
